```python
def display_multiplication_table(number):
    """
    Displays the multiplication table for a given number.
    """
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11): # Generates table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
display_multiplication_table(num)

print("\n") # Add a newline for better readability between examples

num = 12
display_multiplication_table(num)
```

```python
def factorial(n):
    """
    Calculates the factorial of a non-negative integer n.
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")

number = 0
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")

number = -3
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def check_odd_even(number):
    """
    Checks if a given number is odd or even.
    """
    if number % 2 == 0:
        print(f"The number {number} is Even.")
    else:
        print(f"The number {number} is Odd.")

# Example usage:
num1 = 7
check_odd_even(num1)

num2 = 10
check_odd_even(num2)

num3 = 0
check_odd_even(num3)

num4 = -5
check_odd_even(num4)
```

```python
def is_palindrome(number):
    """
    Checks if a given number is a palindrome.
    A number is a palindrome if it reads the same forwards and backwards.
    """
    # Convert the number to a string to easily reverse it
    num_str = str(number)

    # Compare the original string with its reversed version
    if num_str == num_str[::-1]:
        print(f"The number {number} is a Palindrome.")
        return True
    else:
        print(f"The number {number} is not a Palindrome.")
        return False

# Example usage:
num1 = 121
is_palindrome(num1)

num2 = 12321
is_palindrome(num2)

num3 = 12345
is_palindrome(num3)

num4 = -101 # Negative numbers are generally not considered palindromes in typical definitions
is_palindrome(num4)
```

# Task
Develop an expert system with defined rules, a Python-based rule engine, and a text-based user interface to provide recommendations based on user input, demonstrating its decision-making capabilities.

## Define Expert System Rules

### Subtask:
Define a set of simple 'if-then' rules for a specific decision-making domain, such as recommending an activity based on weather conditions.


### Chosen Domain: Activity Recommendations Based on Weather Conditions

For this expert system, we will focus on recommending outdoor activities based on current weather conditions. This is a common and intuitive scenario for demonstrating 'if-then' rules.

#### Rules will consider:
*   **Weather type:** Sunny, Rainy, Cloudy
*   **Temperature:** Hot, Moderate, Cold

Based on these conditions, the system will suggest suitable activities.

**Reasoning**:
Now that the domain is established, I will define the 'if-then' rules for the expert system and store them in a Python list of dictionaries.



```python
expert_rules = [
    {
        "condition": {"weather": "Sunny", "temperature": "Hot"},
        "recommendation": "Go swimming or stay indoors with AC."
    },
    {
        "condition": {"weather": "Sunny", "temperature": "Moderate"},
        "recommendation": "Go for a hike or a bike ride."
    },
    {
        "condition": {"weather": "Sunny", "temperature": "Cold"},
        "recommendation": "Enjoy a brisk walk or outdoor sports."
    },
    {
        "condition": {"weather": "Rainy", "temperature": "Any"},
        "recommendation": "Read a book, watch a movie, or visit a museum."
    },
    {
        "condition": {"weather": "Cloudy", "temperature": "Moderate"},
        "recommendation": "Have a picnic or a leisurely walk in the park."
    },
    {
        "condition": {"weather": "Cloudy", "temperature": "Cold"},
        "recommendation": "Visit a coffee shop or explore indoor attractions."
    }
]

print("Expert system rules defined successfully.")
```

## Implement Rule Engine

### Subtask:
Develop Python code to create a basic rule engine that can store the defined rules and infer decisions based on user inputs or facts.


**Reasoning**:
I will define the `recommend_activity` function to implement the rule engine logic as per the instructions, iterating through `expert_rules` to find a matching recommendation based on weather and temperature inputs.



```python
def recommend_activity(weather, temperature):
    """
    Recommends an activity based on weather conditions and temperature
    using the defined expert_rules.
    """
    for rule in expert_rules:
        condition = rule["condition"]
        if condition["weather"] == weather:
            if condition["temperature"] == temperature or condition["temperature"] == "Any":
                return rule["recommendation"]
    return "No specific recommendation can be made for the given conditions. Try different inputs."

print("Rule engine function 'recommend_activity' defined.")
```

**Reasoning**:
Now that the `recommend_activity` function has been defined, I will add example usage to demonstrate its functionality and verify that it correctly infers recommendations based on various weather and temperature inputs.



```python
print("\n--- Testing recommend_activity function ---")

# Test Case 1: Sunny and Hot
weather_input = "Sunny"
temperature_input = "Hot"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")

# Test Case 2: Sunny and Moderate
weather_input = "Sunny"
temperature_input = "Moderate"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")

# Test Case 3: Rainy (Any temperature)
weather_input = "Rainy"
temperature_input = "Cold"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")

# Test Case 4: Cloudy and Moderate
weather_input = "Cloudy"
temperature_input = "Moderate"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")

# Test Case 5: No matching rule (e.g., Snowy)
weather_input = "Snowy"
temperature_input = "Cold"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")

# Test Case 6: No matching rule (e.g., Sunny and Extreme)
weather_input = "Sunny"
temperature_input = "Extreme"
recommendation = recommend_activity(weather_input, temperature_input)
print(f"Weather: {weather_input}, Temperature: {temperature_input} -> Recommendation: {recommendation}")
```

## Create User Interface

### Subtask:
Design a simple text-based interface to allow users to input facts (e.g., current weather) and interact with the expert system to get recommendations.


---

**Reasoning**:
To create a user-friendly interface, I will now implement a text-based input system that prompts the user for weather and temperature conditions. These inputs will then be passed to the `recommend_activity` function, and the resulting recommendation will be displayed to the user.

---

**Reasoning**:
I will implement a text-based user interface by prompting the user for weather and temperature conditions, then use these inputs to call the `recommend_activity` function, and finally print the resulting recommendation.



```python
print("\n--- Expert System User Interface ---")
print("Please provide the current weather conditions to get an activity recommendation.")

# 1. Prompt the user for weather condition
weather_input = input("Enter weather (Sunny, Rainy, Cloudy): ").strip().capitalize()

# 2. Prompt the user for temperature condition
temperature_input = input("Enter temperature (Hot, Moderate, Cold): ").strip().capitalize()

# Validate inputs against defined conditions (optional but good practice)
valid_weathers = ["Sunny", "Rainy", "Cloudy"]
valid_temperatures = ["Hot", "Moderate", "Cold", "Any"] # "Any" is for internal rule logic, not usually user input

if weather_input not in valid_weathers:
    print(f"Invalid weather input: '{weather_input}'. Please choose from {', '.join(valid_weathers)}.")
elif temperature_input not in valid_temperatures[:-1]: # Exclude "Any" from user input validation
    print(f"Invalid temperature input: '{temperature_input}'. Please choose from {', '.join(valid_temperatures[:-1])}.")
else:
    # 3. Call the recommend_activity function with the user's inputs
    recommendation = recommend_activity(weather_input, temperature_input)

    # 4. Print the recommendation to the user
    print(f"\nBased on your input:\nWeather: {weather_input}\nTemperature: {temperature_input}\nRecommendation: {recommendation}")
```
